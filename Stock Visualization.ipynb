```python
import pandas as pd
import matplotlib.pyplot as plt
import os, sys
```

```python
path = "stock_csvs\\"
dirs = os.listdir(path)
```

```python
i = 0

for file in dirs:
    df = pd.read_csv(path + file, sep=',')
    df.plot.line(x='Date', y='Close', subplots=True)
    plt.title(df['Company'][i] + " 1 year stock")
    i += 1
```
